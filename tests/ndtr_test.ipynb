```python
import pytensor_special as pts
import pymc as pm
import arviz as az
```

```python
#Test ndtr by generating uniform RV via normal distribution
with pm.Model() as model:
    norm = pm.Normal('norm', mu=1, sigma=3)
    uniform = pm.Deterministic('uniform', pts.ndtr(1, 3, norm))
    idata = pm.sample_prior_predictive(10_000)

az.plot_posterior(idata, var_names=['uniform'], group="prior")
```
